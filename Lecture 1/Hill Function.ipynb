```python
import scipy.integrate as integrate
import matplotlib.pyplot as pyplot
import numpy as numpy
```

## Michaelis - Menten Equation

$V = \frac{dX}{dt} = \frac{V_{max}[S]}{K_m+[S]}$


```python
def M_M(s, params):
    return params['Vmax']*s / (params['Km'] + s)
```

```python
sample_params = {'Vmax': 1.0, 'Km': 1.0}
x = numpy.linspace(0,2,1000)
y = [M_M(i,sample_params) for i in x]
```

```python
pyplot.plot(x,y)
```

## Hill Equation

### Hill Activation

$V = \frac{dX}{dt} = \frac{V_{max}[S]^n}{K_{D}^n+[S]^n}$


```python
def hill_act(s, n, params):
    return params['Vmax']*(s ** n) / (params['K'] ** n + s ** n)
```

```python
sample_params = {'Vmax': 1.0, 'K': 1.0}
x = numpy.linspace(0,2,1000)
y = [(j,[hill_act(i,j,sample_params) for i in x]) for j in range(1,10,2)]
```

```python
for p in y:
    pyplot.plot(x,p[1],label="n=" + str(p[0]))
pyplot.legend(loc=2)
```

### Hill Repression

$V = \frac{dX}{dt} = \frac{V_{max}}{1+\left(\frac{[S]}{K_D}\right)^n}$


```python
def hill_rep(s, n, params):
    return params['Vmax'] / (1+ (s/params['K']) ** n )
```

```python
sample_params = {'Vmax': 1.0, 'K': 1.0}
x = numpy.linspace(0,2,1000)
y = [(j,[hill_rep(i,j,sample_params) for i in x]) for j in range(1,10,2)]
```

```python
for p in y:
    pyplot.plot(x,p[1],label="n=" + str(p[0]))
pyplot.legend(loc=2)
```
