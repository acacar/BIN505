```python
import scipy.integrate as integrate
import matplotlib.pyplot as pyplot
import numpy as numpy

```

### Rabbits

$\frac{dR}{dt} = \alpha R $

```python
def rabbits(R, t, alpha, beta):
    return  alpha*R 
```

```python
initial = 2.0   # start with two rabbits
t = numpy.linspace(start=0,stop=10,num=1000) # four years
```

```python
result = integrate.odeint(rabbits,initial,t,args=(2.2, 0))
```

```python
fig = pyplot.figure()
ax = fig.add_subplot(111)
pyplot.plot(t,result[:,0],'r',label='Rabbits')
ax.set_xlabel('Years')
ax.set_ylabel('#Rabbits')
pyplot.legend()
```

### Rabbits and Foxes

$\frac{dR}{dt} = \alpha R - \beta RF$


$\frac{dF}{dt} = \delta RF - \gamma F$

```python
def rabbits_and_foxes(populations, t, alpha, beta, delta, gamma):
    R, F = populations
    return  [ alpha*R - beta*R*F, 
             delta*R*F - gamma*F]
```

```python
initial = [2.0,1.0] # Start with 2 rabbits and one fox
t = numpy.linspace(0,10,1000) # ten years
```

```python
result = integrate.odeint(rabbits_and_foxes,initial,t,(5.0,1,0.5,15))
```

```python
fig = pyplot.figure()
ax = fig.add_subplot(111)
pyplot.plot(t,result[:,0],'r',label='Rabbits')
pyplot.plot(t,result[:,1],'b',label='Foxes')
ax.set_xlabel('Years')
ax.set_ylabel('Number of ...')
pyplot.legend()
```
