```python
import numpy as np    # For numerics
import matplotlib.pyplot as plt  # For plotting
import scipy.integrate as spi  #For integration
```

```python
alpha = 5.0
beta = 1.0
def rate_of_rabbits(populations, t):
    R, F = populations
    # Equivalent to saying:
    # R = populations[0]
    # F = populations[1]
    return alpha*R - beta*R*F

gamma = 2.0
delta = 0.5

def rate_of_foxes(populations, t):
    R, F = populations
    return delta*R*F - gamma*F

def rate_of_pop_change(populations, t):
    return [rate_of_rabbits(populations, t),
            rate_of_foxes(populations, t)]
```

```python
rate_of_pop_change([10.2, 20], 875)
```

```python
t = np.linspace(0,14,1000)
```

```python
result = spi.odeint(rate_of_pop_change, [2.0, 1.0], t)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(t, result[:,0], 'r', label="Rabbits")
plt.plot(t, result[:,1], 'b', label="Foxes")
ax.set_ylabel("Animals per hectare")
plt.legend()
```
